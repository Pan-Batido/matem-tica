2. ¿Cuál es la forma algebraica de la función? ¿Qué podrían significar la pendiente y el coeficiente de posición en este caso?


```python
import numpy as np
x = np.array([1,2,3,4,5])
y= np.array([8.5,11,13.5,16,18.5])
pendiente, intercepto = np.polyfit(x,y,1)
print("La pendiente es",round(pendiente,2),"y el coeficiente de posición es",round(intercepto,2))
```

$m=2.5$ Significa que por cada hoja impresa el tiempo aumenta en 2.5 segundos. 

$n=6$ Significa que la impresora tarda 6 segundos en comenzar a imprimir.

La función es: $T(x)=2.5x+6$

3. Calcule el tiempo que tarda la impresora en procesar 17 hojas.


```python
def T(x):
    return 2.5*x+6
```

```python
a=17
print("Al imprimir",a,"hojas la impresora tarda",T(a),"segundos")
```

4. Si la impresora tardó 2 minutos y 6 segundos, ¿Cuántas hojas imprimió?


2 minutos y 6 segundos = 126 segundos.

Buscamos $x$ que resuelva $2.5x+6=126$

```python
from scipy.optimize import fsolve
def F(x):
    return T(x)-126
xo = np.linspace(0,100,100)
soluciones = np.around(fsolve(F, xo),1)
soluciones_unicas = np.unique(soluciones)
print(soluciones_unicas)
```

La impresora tarda 2 minutos y 6 segundos en imprimir 48 hojas.

5. El jefe nos indica que hay otra impresora que podría ser conveniente. Ésta tarda 1,5 segundos por hoja, pero toma un tiempo de 25 segundos en
prenderse. ¿Para cuántas hojas es más conveniente esta segunda impresora?

Esta impresora tiene asociada la función de tiempo $G(x)=1.5x+25$

```python
def G(x):
    return 1.5*x+25
```

```python
import matplotlib.pyplot as plt
x = np.arange(0,41,1)
plt.plot(x, T(x), label = 'Impresora 1')
plt.plot(x, G(x), label = 'Impresora 2')
plt.axis([0,40,0,100])
plt.title('Nombre del gráfico')
plt.xlabel('Nombre eje x')
plt.ylabel('Nombre eje y')
plt.grid(True)
plt.legend()
plt.show()
```

```python
def F(x):
    return T(x)-G(x)
xo = np.linspace(0,40,100)
soluciones = np.around(fsolve(F, xo),1)
soluciones_unicas = np.unique(soluciones)
print(soluciones_unicas)
```

Para 19 hojas los tiempos son iguales. Si se imprimen entre 1 y 18 hojas, conviene la impresora 1, pero si se imprimen desde 20 hojas, conviene la impresora 2.
